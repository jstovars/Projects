```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlencode

def bing_search(query):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
    }
    params = {
        'q': query,
        'count': '1'  # N√∫mero de resultados
    }
    response = requests.get("https://www.bing.com/search?" + urlencode(params), headers=headers)
    if response.status_code != 200:
        print(f"Failed to fetch results. Status code: {response.status_code}")
        return []

    soup = BeautifulSoup(response.text, 'html.parser')
    results = []
    for h in soup.find_all('h2'):
        link = h.find('a', href=True)
        if link:
            results.append(link['href'])
    
    if not results:
        print("No links found. Make sure the HTML tags have not changed.")
    
    return results

# Ejemplo de uso
query = "que incluir en mi CV"
links = bing_search(query)
for idx, link in enumerate(links):
    print(f"Link {idx+1}: {link}")
```

```python
import requests
from bs4 import BeautifulSoup

def search_bing(query, num_results=5):
    # Bing search URL
    url = 'https://www.bing.com/search'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    params = {'q': query}
    
    response = requests.get(url, headers=headers, params=params)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    results = []
    for item in soup.find_all('li', {'class': 'b_algo'}):
        link = item.find('a')['href']
        results.append(link)
        if len(results) == num_results:
            break
    
    return results

def scrape_content(urls):
    headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:88.0) Gecko/20100101 Firefox/88.0'}
    contents = []
    
    for url in urls:
        print(f"Scraping {url}")
        try:
            response = requests.get(url, headers=headers)
            soup = BeautifulSoup(response.content, 'html.parser')
            
            text_content = {'url': url, 'title': None, 'paragraphs': []}
            
            # Extract title
            if soup.title:
                text_content['title'] = soup.title.text.strip()
            
            # Extract paragraphs
            for p in soup.find_all('p'):
                text_content['paragraphs'].append(p.get_text(strip=True))
            
            contents.append(text_content)
        except Exception as e:
            print(f"Failed to scrape {url} due to {e}")
    
    return contents

# Main function to execute the search and scraping
if __name__ == "__main__":
    search_query = "que incluir en mi CV"  # Cambia la consulta seg√∫n sea necesario
    urls = search_bing(search_query)
    scraped_data = scrape_content(urls)
    for data in scraped_data:
        print(f"URL: {data['url']}")
        print(f"Title: {data['title']}")
        print("Paragraphs:")
        for paragraph in data['paragraphs']:
            print(paragraph)

```

```python
def save_to_text_file(data, filename):
    with open(filename, 'w', encoding='utf-8') as file:
        for item in data:
            file.write(f"URL: {item['url']}\n")
            file.write(f"Title: {item['title']}\n")
            file.write("Paragraphs:\n")
            for paragraph in item['paragraphs']:
                file.write(paragraph + '\n')
            file.write('\n')  # L√≠nea en blanco entre cada conjunto de datos

# Llamar a la funci√≥n para guardar los datos en un archivo de texto
if __name__ == "__main__":
    search_query = "que incluir en mi CV"  # Cambia la consulta seg√∫n sea necesario
    urls = search_bing(search_query)
    scraped_data = scrape_content(urls)
    
    # Nombre del archivo de texto
    filename = "guiaCV.txt"
    
    # Guardar los datos en el archivo de texto
    save_to_text_file(scraped_data, filename)
    print(f"Los datos se han guardado en '{filename}'.")
```

```python
from docx import Document

def save_to_word_file(data, filename):
    doc = Document()
    for item in data:
        doc.add_heading('URL:', level=1)
        doc.add_paragraph(item['url'])
        
        doc.add_heading('Title:', level=1)
        doc.add_paragraph(item['title'])
        
        doc.add_heading('Paragraphs:', level=1)
        for paragraph in item['paragraphs']:
            doc.add_paragraph(paragraph)
        
        doc.add_page_break()  # P√°gina nueva entre cada conjunto de datos
    
    doc.save(filename)

# Llamar a la funci√≥n para guardar los datos en un archivo de Word
if __name__ == "__main__":
    search_query = "que incluir en mi CV"  # Cambia la consulta seg√∫n sea necesario
    urls = search_bing(search_query)
    scraped_data = scrape_content(urls)
    
    # Nombre del archivo de Word
    filename = "guiaCV.docx"
    
    # Guardar los datos en el archivo de Word
    save_to_word_file(scraped_data, filename)
    print(f"Los datos se han guardado en '{filename}'.")

```

```python
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

def save_to_pdf(data, filename):
    c = canvas.Canvas(filename, pagesize=letter)
    y = 750  # Posici√≥n inicial en la p√°gina
    
    for item in data:
        c.drawString(100, y, f"URL: {item['url']}")
        y -= 20
        
        c.drawString(100, y, f"Title: {item['title']}")
        y -= 20
        
        c.drawString(100, y, "Paragraphs:")
        y -= 20
        for paragraph in item['paragraphs']:
            c.drawString(120, y, paragraph)
            y -= 20
        
        y -= 20  # Espacio entre cada conjunto de datos
    
        if y < 50:  # Si se alcanza el final de la p√°gina, crear una nueva p√°gina
            c.showPage()
            y = 750  # Reiniciar la posici√≥n en la nueva p√°gina
    
    c.save()

# Llamar a la funci√≥n para guardar los datos en un archivo PDF
if __name__ == "__main__":
    search_query = "que incluir en mi CV"  # Cambia la consulta seg√∫n sea necesario
    urls = search_bing(search_query)
    scraped_data = scrape_content(urls)
    
    # Nombre del archivo PDF
    filename = "guiaCV.pdf"
    
    # Guardar los datos en el archivo PDF
    save_to_pdf(scraped_data, filename)
    print(f"Los datos se han guardado en '{filename}'.")
```
