```python
import PyPDF2

def extract_text_from_pdf(file_path):
    text2 = ""
    with open(file_path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            text2 += page.extract_text() + "\n"  # Agregar un salto de línea entre cada página
    return text2

# Llama a la función con la ruta del archivo PDF
text2 = extract_text_from_pdf("scraped_data_BD.pdf")
print(text2)
```

```python
from llama_index.core import Document

text1 = [Document(id=1, text=text2)]
# Crear un índice de los documentos
index = VectorStoreIndex.from_documents(text1)
```
