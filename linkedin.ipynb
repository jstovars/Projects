```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
# Especifica la ruta al ejecutable de GeckoDriver
geckodriver_path = '"geckodriver_path = '"C:\\Users\\USUARIO\\Downloads\\geckodriver-v0.34.0-win32\\geckodriver"'"'
driver = webdriver.Firefox()
driver.get('https://www.linkedin.com')


# Iniciar sesión en LinkedIn
username = "darkboomb@gmail.com"
password = "12345abcd"

# Esperar a que aparezca el formulario de inicio de sesión
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.NAME, 'session_key')))
driver.find_element(By.NAME, 'session_key').send_keys(username)
driver.find_element(By.NAME, 'session_password').send_keys(password)
driver.find_element(By.XPATH, '//button[@type="submit"]').click()

    # Realizar una búsqueda
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//input[@aria-label="Buscar"]')))
search_input = driver.find_element(By.XPATH, '//input[@aria-label="Buscar"]')
search_input.send_keys("Ingeniería Electrónica Redes")
search_input.send_keys(Keys.RETURN)

    # Hacer clic en el botón "Empleos"
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//button[contains(., "Empleos")]')))
jobs_button = driver.find_element(By.XPATH, '//button[contains(., "Empleos")]')
jobs_button.click()

time.sleep(5)  # Ajusta este tiempo según la velocidad de carga de la página

# Obtener el HTML de la página actual
page_source = driver.page_source

    # Analizar el HTML con BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

    # Extraer todo el texto del HTML
all_text = soup.get_text(separator="\n")

    # Imprimir el texto extraído
print(all_text)


driver.quit()
```

```python

# Encuentra las posiciones de las palabras clave
pos_about_job = all_text.find("Acerca del empleo")
pos_requirements = all_text.find("Requisitos")
pos_about_company = all_text.find("Acerca de la empresa")

# Extrae el contenido entre las palabras clave
content_about_job = all_text[pos_about_job:pos_requirements].strip()
content_requirements = all_text[pos_requirements:pos_about_company].strip()
content_about_company = all_text[pos_about_company:].strip()

# Concatena el contenido en una nueva cadena
final_content = content_about_job + content_requirements + content_about_company

print(final_content)
```

```python
import re

# Reemplazar múltiples espacios en blanco por un salto de línea
final_content = re.sub(r'\s{2,}', '\n', final_content)
final_content = final_content.replace("Mostrar más", "")
final_content = final_content.replace("mostrar más", "")
print(final_content)
print(final_content)
```
